```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.preprocessing import StandardScaler
from sklearn.impute import SimpleImputer
```

```python
test_data = pd.read_csv('/content/test.csv')
train_data = pd.read_csv('/content/train (1).csv')
```

```python
test_data.isnull().sum()
```

```python
train_data.isnull().sum()
```

```python
train_data.shape
```

```python
train_data['TotalBath'] = train_data['FullBath'] + train_data['HalfBath'] * 0.5 + train_data['BsmtFullBath'] + train_data['BsmtHalfBath'] * 0.5
X_Train = train_data[['GrLivArea', 'BedroomAbvGr', 'TotalBath']]
Y_Train = train_data['SalePrice']
```

```python
imputer = SimpleImputer(strategy='mean')
X_Train_imputed = imputer.fit_transform(X_Train)
```

```python
scaler = StandardScaler()
```

```python
X_Train_scaled = scaler.fit_transform(X_Train_imputed)
```

```python
test_data['TotalBath'] = test_data['FullBath'] + test_data['HalfBath'] * 0.5 + test_data['BsmtFullBath'] + test_data['BsmtHalfBath'] * 0.5
X_Test = test_data[['GrLivArea', 'BedroomAbvGr', 'TotalBath']]
X_Test_imputed = imputer.transform(X_Test)
X_Test_scaled = scaler.transform(X_Test_imputed)
```

```python
model = LinearRegression()
```

```python
model.fit(X_Train_scaled,Y_Train)
```

```python
X_Train_prediction = model.predict(X_Train_scaled)
```

```python
mse = mean_squared_error(X_Train_prediction,Y_Train)
mse
```

```python
Y_Test_pred = model.predict(X_Test_scaled)
```

```python
submission = pd.DataFrame({
    'Id': test_data['Id'],
    'SalePrice': Y_Test_pred
})
submission.to_csv('my_submission.csv', index = False)
print('Submission Saved')
```
